# Data Cleaning

```python
df.info()
```

```python
df.drop(columns=['channel'],inplace=True)
```

```python
df.columns
```

```python
df['keywords'] = df['keywords'].fillna("Not Avilable")
```

```python
df['keywords'].isnull().sum()
```

```python
df['published_at'] = pd.to_datetime(df['published_at'])
```

```python
df['published_at'].dtypes
```

```python
df['collection_date'] = pd.to_datetime(df['collection_date'],dayfirst=True)
```

```python
df['collection_date'].dtypes
```

```python
df.dtypes
```

```python
df.sample()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
# clean text col 
text_col = ['title','channel_name','tags','keywords']

for col in text_col: 
    df[col] = df[col].str.strip()
```

```python
# final data validation 
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.to_csv("../data/cleaned/youtube_clean.csv",index=False)
print('save')
```
